```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el dataset directamente desde la librería
df = sns.load_dataset('penguins')

# Visualización inicial de la estructura
print("--- Estructura del Dataset ---")
print(df.info())
print("\n--- Primeras 5 filas ---")
print(df.head())
```

```python
# Contar valores nulos por columna
null_counts = df.isnull().sum()
print("Valores faltantes por columna:\n", null_counts)

# Visualizar la ubicación de los nulos
plt.figure(figsize=(10, 4))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title("Mapa de calor de datos faltantes")
plt.show()
```

```python
# Análisis multivariante segmentado por especie
sns.pairplot(df, hue='species', markers=["o", "s", "D"])
plt.suptitle("Relaciones entre variables físicas por Especie", y=1.02)
plt.show()
```

```python
plt.figure(figsize=(12, 5))

# Relación Especie vs Isla
plt.subplot(1, 2, 1)
sns.countplot(data=df, x='island', hue='species')
plt.title("Distribución de Especies por Isla")

# Relación Especie vs Sexo
plt.subplot(1, 2, 2)
sns.countplot(data=df, x='sex', hue='species')
plt.title("Distribución de Especies por Sexo")

plt.tight_layout()
plt.show()
```
